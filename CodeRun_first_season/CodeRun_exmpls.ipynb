## 156. Опечатки (medium)
https://coderun.yandex.ru/problem/typos?currentPage=1&pageSize=10&tag=first_2023_ml&rowNumber=4

```python
from collections import deque

with open('input.txt', 'r') as f:
    data = f.read().splitlines()
    
mean = data[0]
inp = data[1]
n = int(data[2])
dic_of_correct_replaces_for_each_mistake, answer = {}, -1

for indx in range(3, n + 3):
    true_letters, mistake = data[indx].split()
    if mistake not in dic_of_correct_replaces_for_each_mistake:
        dic_of_correct_replaces_for_each_mistake[mistake] = []
    dic_of_correct_replaces_for_each_mistake[mistake].append(true_letters)

deck = deque()
deck.append([inp, 0])

while deck:
    
    node, node_count = deck.popleft()
    if node_count > 4:
        break
    if node == mean:
        answer = node_count
        break
        
    for mistake in dic_of_correct_replaces_for_each_mistake:
        if mistake in node:
            for possible_replace in dic_of_correct_replaces_for_each_mistake[mistake]:
                leng = len(possible_replace)
                for left in range(len(node) - leng + 1):
                    if possible_replace == mean[left : leng + left]:
                        deck.append([node[:left] + possible_replace + node[leng + left:], node_count + 1])

print(answer)
```

## 262. Лента рекомендаций (hard)

https://coderun.yandex.ru/problem/recommendations?currentPage=1&difficulty=HARD&pageSize=10&tag=first_2023_ml&rowNumber=4

```python
import heapq

n, m = list(map(int, input().split()))
counter, graph = 0, {}

while n != 0:
    n -= 1
    type_ = int(input())    
    if type_ not in graph:
        graph[type_] = []
    graph[type_].append(counter)
    counter += 1

heap = [list(tup) for tup in list(zip(graph.values(),graph.keys()))]
heapq.heapify(heap)

prev, answer, stack = None, [], []
while True: 
    
    if heap and heap[0][1] != prev:
        cur = heapq.heappop(heap)
        ind = cur[0].pop(0)
        prev = cur[1]
        answer.append(ind)
        if cur[0]:
            heapq.heappush(heap, cur)
    
    elif len(heap) > 1:
        prevv = heapq.heappop(heap)
        cur = heapq.heappop(heap)
        ind = cur[0].pop(0)
        prev = cur[1]
        answer.append(ind)
        if cur[0]:
            heapq.heappush(heap, cur)
        heapq.heappush(heap, prevv)
    else:
        break

print(*answer)
```

## 243. Рекомендательная система SVD (medium)
https://coderun.yandex.ru/problem/svd-recommender?currentPage=2&difficulty=MEDIUM&pageSize=10&tag=first_2023_ml&rowNumber=14

```python
import numpy as np

with open('input.txt', 'r') as f:
    data = f.read().splitlines()
k, U, M, D, T = map(int, data[0].split())

matrix, train_r = [[-1] * (M + 1) for _ in range(U + 1)], []

for sub_data in data[1:D + 1]:
    sub_data = list(map(float, sub_data.split()))
    matrix[sub_data[0]][sub_data[1]] = sub_data[2]
    train_r.append(sub_data[2])

train_r = np.array(train_r)

n_factors = 10
n_epochs = 15
learning_rate = .01
regularization = .02

n_users = U
n_items = M
global_mean = train_r.mean()

bu = np.zeros(n_users, np.double)
bi = np.zeros(n_items, np.double)
pu = np.random.normal(0, 0.1, (n_users, n_factors))
qi = np.random.normal(0, 0.1, (n_items, n_factors))

for current_epoch in range(n_epochs):
    for indx_u in range(1, n_users + 1):
        for indx_m in range(1, n_items + 1):
            if matrix[indx_u][indx_m] == -1:
                continue
                
            err = matrix[indx_u][indx_m] - (bu[indx_u - 1] + bi[indx_m - 1] +
                                            global_mean + qi[indx_m - 1] @ pu[indx_u - 1])

            bu[indx_u - 1] += learning_rate * (err - regularization * bu[indx_u - 1])
            bi[indx_m - 1] += learning_rate * (err - regularization * bi[indx_m - 1])

            pu[indx_u - 1] += learning_rate * (err * qi[indx_m - 1] - regularization * pu[indx_u - 1])
            qi[indx_m - 1] += learning_rate * (err * pu[indx_u - 1] - regularization * qi[indx_m -1])

lower_bound, higher_bound = train_r.min(), train_r.max()

for sub_data in data[D + 1:]:
    test_u, test_m = map(float, sub_data.split())
    test_u -= 1
    test_m -= 1
    
    est = global_mean + bu[test_u] + bi[test_m] + pu[test_u] @ qi[test_m]
    est = min(higher_bound, est)
    est = max(lower_bound, est)
    print(est)
```

## 301. Бутылки рома (medium)
https://coderun.yandex.ru/problem/bottles-of-rum?currentPage=2&difficulty=MEDIUM&pageSize=10&tag=first_2023_ml&rowNumber=18

```python
from math import factorial
import numpy as np

string, dic = input(), {}
for letter in string:
    dic[letter] = dic.get(letter, 0) + 1

stack, answer, visited = [list(dic.values())], 0, set()
while stack:
    
    node = stack.pop()
    to_check = tuple(node)
    if to_check in visited:
        continue
        
    visited.add(to_check)
    len_node = len(node)
    to_add = factorial(sum(node)) / np.prod([factorial(num) for num in node])

    answer += to_add
    
    for count_indx in range(len_node):
        if node[count_indx] > 0:
            node[count_indx] -= 1
            to_continue_lst = node.copy()
            if sum(to_continue_lst) > 0:
                stack.append(to_continue_lst)
            node[count_indx] += 1
print(answer)
```
